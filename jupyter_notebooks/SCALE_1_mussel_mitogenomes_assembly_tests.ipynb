```python
import subprocess
```

```python
for sra in ["SRR31437301", "SRR27608898", "SRR12267064"]:
    task = "sra_dump"
    mem = "32"
    cpus = "4"
    prefix = sra
    outpath = "~/scratch/mussels/"
    subprocess.run(["sbatch --mem=" + mem + "g --ntasks=" + 
         cpus + " ../" + task + ".sh " + 
         prefix + " " + outpath], shell=True)
```
