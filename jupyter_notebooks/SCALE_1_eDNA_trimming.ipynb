We seem to have some specific tiles that are bad for MiFish - perhaps from overloading? We'll use BBmap's filterbytile (which does not bias quality scores as it removes problematic tiles rather than problematic reads or parts of reads) as a start, then use cutadapt to trim, and then process with dada2:

```python
%%bash
cat ~/scratch/reads/*_R1.fastq.gz > ~/scratch/reads/all_1.fq.gz
cat ~/scratch/reads/*_R2.fastq.gz > ~/scratch/reads/all_2.fq.gz
```

```python
%%bash
source ~/.bashrc
cd ~/scratch/reads/
conda activate assembly
fastqc ~/scratch/reads/all_1.fq.gz ; fastqc ~/scratch/reads/all_2.fq.gz
```

```python
%%bash
source ~/.bashrc
conda activate ipyrad
filterbytile.sh \
    in1=~/scratch/reads/all_1.fq.gz \
    in2=~/scratch/reads/all_2.fq.gz \
    dump=~/scratch/reads/dump.flowcell
```

```python
%%bash
source ~/.bashrc
conda activate ipyrad
mkdir -p ~/scratch/reads/filterbytile
for samp in $( ls ~/scratch/reads/*_R1.fastq.gz | rev | cut -d"_" -f2- | rev | sed "s:/storage/home/nsl5160/scratch/reads/::g" ); do
    filterbytile.sh \
        in=~/scratch/reads/${samp}_R1.fastq.gz \
        in=~/scratch/reads/${samp}_R2.fastq.gz \
        out1=~/scratch/reads/filterbytile/${samp}_R1.fastq.gz \
        out2=~/scratch/reads/filterbytile/${samp}_R2.fastq.gz \
        indump=~/scratch/reads/dump.flowcell
done
```

```python
%%bash
source ~/.bashrc
cd /storage/home/nsl5160/scratch/reads/filterbytile
mkdir -p ./trimmed
conda activate assembly
for i in $( ls ~/scratch/reads/*_R1.fastq.gz | rev | cut -d"_" -f2- | rev ) ; do
    PREFIX=$( echo ${i} | cut -d"_" -f1 | sed "s:/storage/home/nsl5160/scratch/reads/::g" )
    cutadapt -g ^GTCGGTAAAACTCGTGCCAGC -G ^CATAGTGGGGTATCTAATCCCAGTTTG \
        -j 8 -e 0.4 --discard-untrimmed \
        -o ./trimmed/${PREFIX}_R1.fq.gz -p ./trimmed/${PREFIX}_R2.fq.gz \
        ${i}_R1.fastq.gz ${i}_R2.fastq.gz
done
```

```python
%%bash
source ~/.bashrc
cd /storage/home/nsl5160/scratch/reads/filterbytile/trimmed
conda activate assembly
for i in $( ls . ) ; do
    fastqc ${i}
done
```

We'll repeat the trimming for the mussel data, which uses primers from: https://onlinelibrary.wiley.com/doi/full/10.1002/edn3.390. The errors look fairly tame and no tile-specific issues were present with the quality scores, so we'll skip that step and go straight into trimming with cutadapt:

```python
%%bash
source ~/.bashrc
cd /storage/home/nsl5160/scratch/reads/inverts/mussel.reads/
mkdir -p ./trimmed
conda activate assembly
for i in $( ls /storage/home/nsl5160/scratch/reads/inverts/mussel.reads/*_R1.fastq.gz | rev | cut -d"_" -f2- | rev ) ; do
    PREFIX=$( echo ${i} | cut -d"_" -f2 )
    cutadapt -g ^AGNCTTYTVATYCGDGCYGA -G ^CCRGTHCCNACACCHCTCTC \
        -j 8 -e 0.3 --discard-untrimmed \
        -o ./trimmed/${PREFIX}_R1.fq.gz -p ./trimmed/${PREFIX}_R2.fq.gz \
        ${i}_R1.fastq.gz ${i}_R2.fastq.gz
done
```

Our reads have really suboptimal error profiles,  potentially resulting from insufficient PhiX or from overload of the flow cells. In an open issue on the DADA2 GitHub, the authors recommend merging reads and processing with DADA2 using the single end mode. We will generate a merged dataset to also try that option here:

```python
%%bash
source ~/.bashrc
conda activate ipyrad
cd /storage/home/nsl5160/scratch/reads/filterbytile/trimmed/
mkdir -p ./merged
for SAMP in $( ! ls /storage/home/nsl5160/scratch/reads/filterbytile/trimmed/*_R1.fq.gz | rev | cut -d"_" -f2- | cut -d"/" -f1 | rev ); do
    vsearch \
        --threads 8 \
        --fastq_mergepairs ${SAMP}_R1.fq.gz \
        --reverse ${SAMP}_R2.fq.gz \
        --fastqout ./merged/${SAMP}.fq \
        --fastq_maxdiffpct 20 \
        --fastq_minovlen 20 \
        --fastq_maxmergelen 200 \
        --fastq_minmergelen 140 \
        --fastq_truncqual 0 
done
```
