```python
import glob
import subprocess
```

```python
files = glob.glob("/storage/home/nsl5160/scratch/brook_trout/raw/01.RawData/*/*_2.fq.gz")
```

```python
for file in files:
    task = "trimgalore"
    mem = "24"
    cpus = "8"
    inprefix = file.rsplit("_", 1)[0]
    r1 = inprefix + "_1.fq.gz"
    r2 = inprefix + "_2.fq.gz"
    outpath = "/storage/home/nsl5160/scratch/brook_trout/trimmed/"
    outprefix = file.split("/", 9)[-1].rsplit("_", 4)[0]
    subprocess.run(["sbatch --mem=" + mem + "g --ntasks=" + 
         cpus + " ../" + task + ".sh " + 
         r1 + " " + r2 + " " + 
         outpath + " " + outprefix], shell=True)
```
